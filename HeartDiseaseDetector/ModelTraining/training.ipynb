# Heart Disease SVM Training Notebook

This notebook mirrors the logic in `ModelTraining/svm.py` and is intended
for interactive experimentation with the UCI Heart Disease dataset (`heart.csv`).

Steps you can run here:
1. Load `heart.csv` into a pandas DataFrame.
2. Explore feature distributions and correlations.
3. Split the data into train/test sets.
4. Scale features using `StandardScaler`.
5. Train an SVM classifier and evaluate performance.
6. Persist the trained model and scaler with `joblib.dump` so the Django
   app can load and use them for web-based predictions.

> Note: The Django app expects the trained artefacts at
> `ModelTraining/heart_svm_model.pkl` with the dictionary structure
> `{ "model": clf, "scaler": scaler, "features": feature_names }`.
